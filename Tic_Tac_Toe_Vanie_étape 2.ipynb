```python
import tkinter as tk
from tkinter import messagebox

# Initialisation de la fenêtre
root = tk.Tk()
root.title("TIC-TAC-TOE")

# Variables globales
digits = [1, 2, 3, 4, 5, 6, 7, 8, 9]  # Représente les cases disponibles
count = 0  # Compteur des mouvements
mark = ""  # Marque du joueur actuel, X ou O
panels = ["panel", "", "", "", "", "", "", "", "", ""]  # État du plateau de jeu
buttons = []  # Liste pour stocker les boutons

# Fonction de vérification des conditions de victoire
def win():
    # Combinaisons gagnantes possibles
    winning_combinations = [(1, 2, 3), (4, 5, 6), (7, 8, 9),  # Lignes horizontales
                            (1, 4, 7), (2, 5, 8), (3, 6, 9),  # Lignes verticales
                            (1, 5, 9), (3, 5, 7)]  # Diagonales
    
    for combo in winning_combinations:
        if panels[combo[0]] == panels[combo[1]] == panels[combo[2]] and panels[combo[0]] != "":
            return True
    return False

# Fonction pour gérer les mouvements et les mises à jour
def checker(digit):
    global count, mark, digits, panels

    if digit in digits:  # Si le numéro est valide
        digits.remove(digit)  # Supprimer le digit des cases disponibles
        panels[digit] = mark  # Mettre à jour le panneau avec la marque du joueur
        
        # Mise à jour du bouton correspondant
        buttons[digit].config(text=mark, state="disabled")
        
        # Incrémenter le nombre de coups
        count += 1
        
        # Vérifier si un joueur a gagné
        if win():
            messagebox.showinfo("Félicitations", f"{mark} a gagné!")
            root.destroy()  # Fermer la fenêtre du jeu
            return

        # Si le jeu est terminé
        if count > 8:
            messagebox.showinfo("Match Tied", "Le match est terminé avec égalité.")
            root.destroy()  # Fermer la fenêtre du jeu
            return

        # Changer de joueur
        mark = "X" if mark == "O" else "O"

# Fonction pour créer le plateau de jeu
def create_board():
    global mark
    mark = "X"  # Le premier joueur commence avec "X"

    # Créer les boutons pour chaque case du jeu et les ajouter à la liste
    for i in range(1, 10):
        button = tk.Button(root, text=str(i), width=10, height=3, font=("Arial", 20), command=lambda i=i: checker(i))
        button.grid(row=(i-1)//3, column=(i-1)%3)
        buttons.append(button)

# Lancer le jeu
create_board()

# Lancer la boucle principale de l'application
root.mainloop()
```
