```python
import tkinter as tk
from tkinter import messagebox

# Créons notre fenêtre principale
root = tk.Tk()
root.title("TIC-TAC-TOE")

# Déclarons nos variables globales
digits = [1, 2, 3, 4, 5, 6, 7, 8, 9]
count = 0
mark = "X"  # Le joueur 1 commence
panels = ["panel", "", "", "", "", "", "", "", "", ""]
buttons = [None] * 10  # Index 0 non utilisé

# Ajoutons des labels pour afficher les symboles des joueurs
label_frame = tk.Frame(root)
label_frame.grid(row=0, column=0, columnspan=3, pady=10)

player1_label = tk.Label(label_frame, text="Joueur 1 : X", font=("Arial", 14), fg="blue")
player1_label.pack(side="left", padx=20)

player2_label = tk.Label(label_frame, text="Joueur 2 : O", font=("Arial", 14), fg="green")
player2_label.pack(side="right", padx=20)

# Définissons les combinaisons gagnantes
def win():
    combos = [(1,2,3), (4,5,6), (7,8,9),  # lignes
              (1,4,7), (2,5,8), (3,6,9),  # colonnes
              (1,5,9), (3,5,7)]           # diagonales
    for a, b, c in combos:
        if panels[a] == panels[b] == panels[c] and panels[a] != "":
            return True
    return False

# Fonction de gestion des clics
def checker(digit):
    global count, mark, digits, panels

    if digit in digits:
        digits.remove(digit)
        panels[digit] = mark
        buttons[digit].config(text=mark, state="disabled")
        count += 1

        if win():
            gagnant = "Joueur 1" if mark == "X" else "Joueur 2"
            messagebox.showinfo("Victoire", f"{gagnant} a gagné !")
            root.destroy()
            return

        if count > 8:
            messagebox.showinfo("Égalité", "Match nul !")
            root.destroy()
            return

        # Alternons entre les joueurs
        mark = "O" if mark == "X" else "X"

# Créons les boutons du plateau
def create_board():
    index = 1
    for row in range(1, 4):
        for col in range(3):
            button = tk.Button(root, text="", width=10, height=3, font=("Arial", 20),
                               command=lambda i=index: checker(i))
            button.grid(row=row, column=col, padx=5, pady=5)
            buttons[index] = button
            index += 1

create_board()
root.mainloop()
```
